```python
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        return f"{self.year} {self.brand} {self.model}"

my_car = Car("Acer", "Ryzen 5 6600", 16)
print(my_car.display_info())
```

```python
class StudentRecord:
    def __init__(self):
        self.__grades = []  # private attribute

    def add_grade(self, grade):
        self.__grades.append(grade)

    def get_grades(self):
        return self.__grades

record = StudentRecord()
record.add_grade(1.50)
record.add_grade(2.75)
record.add_grade(1.75)

print("Grades:", record.get_grades())
```

```python
class Device:
    def __init__(self, device_name):
        self.device_name = device_name

    def power_on(self):
        return f"{self.device_name} is powering on..."

class Smartphone(Device):
    def power_on(self):
        return f"{self.device_name} is booting up. Please wait..."

dev = Device("Generic Device")
phone = Smartphone("Samsung Galaxy")

print(dev.power_on())
print(phone.power_on())
```

```python
class Device:
    def power_on(self):
        return "Device is turning on..."

class Smartphone(Device):
    def power_on(self):
        return "Smartphone is booting up..."

class Laptop(Device):
    def power_on(self):
        return "Laptop is starting Windows..."

def start_device(device):
    print(device.power_on())

phone = Smartphone()
laptop = Laptop()

start_device(phone)
start_device(laptop)
```

```python
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

temp = float(input("Enter temperature in Celsius: "))
fahrenheit = TemperatureConverter.celsius_to_fahrenheit(temp)

print(f"{temp}°C = {fahrenheit}°F")
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

v1 = Vector(6, 7)
v2 = Vector(6, 9)

v_add = v1 + v2
v_sub = v1 - v2

print("Added Vector:", v_add.x, v_add.y)
print("Subtracted Vector:", v_sub.x, v_sub.y)
```

```python
from abc import ABC, abstractmethod
# Part 1: Temperature Converter

class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

temp_c = float(input("Enter temperature in Celsius: "))
temp_f = TemperatureConverter.celsius_to_fahrenheit(temp_c)

print(f"{temp_c}°C = {temp_f}°F")



class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

class Refrigerator(Appliance):
    def turn_on(self):
        return "Refrigerator is now cooling."

ref = Refrigerator()
print(ref.turn_on())
```
