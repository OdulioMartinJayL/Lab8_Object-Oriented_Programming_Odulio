```python
class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

    def borrow(self):
        if not self.available:
            return f"❌ '{self.title}' is already borrowed."
        self.available = False
        return f"✅ You borrowed '{self.title}'."

    def return_book(self):
        if self.available:
            return f"❌ '{self.title}' was not borrowed."
        self.available = True
        return f"✅ You returned '{self.title}'."

    def info(self):
        status = "Available" if self.available else "Borrowed"
        return f"{self.title} by {self.author} — {status}"


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, available=True):
        new_book = Book(title, author, available)
        self.books.append(new_book)
        return f"📘 Book '{title}' added."

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book.borrow()
        return f"❌ Book '{title}' not found."

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book.return_book()
        return f"❌ Book '{title}' not found."

    def show_books(self):
        if not self.books:
            return "📚 No books in the library yet."
        return "\n".join(book.info() for book in self.books)


# ---------------------------
# PROGRAM EXECUTION
# ---------------------------

library = Library()

# Add books
num_books = int(input("How many books do you want to add? "))

for _ in range(num_books):
    title = input("Enter book title: ")
    author = input("Enter author: ")
    library.add_book(title, author)

print("\n📚 Current Library Collection:")
print(library.show_books())

# Borrow a book
borrow_title = input("\nEnter a book title to borrow: ")
print(library.borrow_book(borrow_title))

# Return a book
return_title = input("\nEnter a book title to return: ")
print(library.return_book(return_title))

print("\n📚 Updated Library Collection:")
print(library.show_books())
```

```python
class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        if hourly_rate < 0:
            raise ValueError("Hourly rate cannot be negative.")
        if hours_worked < 0:
            raise ValueError("Hours worked cannot be negative.")

        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        base_hours = min(self.hours_worked, 160)
        overtime_hours = max(0, self.hours_worked - 160)

        base_salary = base_hours * self.hourly_rate
        overtime_pay = overtime_hours * (self.hourly_rate * 1.5)
        total_salary = base_salary + overtime_pay

        return base_salary, overtime_pay, total_salary

    def info(self):
        base_salary, overtime_pay, total_salary = self.calculate_salary()

        return (
            f"\nEmployee: {self.name}"
            f"\nTotal Hours Worked: {self.hours_worked}"
            f"\nBase Salary: ₱{base_salary:,.2f}"
            f"\nOvertime Pay: ₱{overtime_pay:,.2f}"
            f"\nTotal Salary: ₱{total_salary:,.2f}"
        )


# ---------------------------
# PROGRAM EXECUTION
# ---------------------------

num_employees = int(input("Enter number of employees: "))

while num_employees <= 0:
    print("❌ You must enter at least ONE employee.")
    num_employees = int(input("Enter number of employees: "))

employees = []

for i in range(num_employees):
    print(f"\n--- Employee {i+1} ---")
    name = input("Enter employee name: ")

    hourly_rate = float(input("Enter hourly rate: "))
    while hourly_rate < 0:
        print("❌ Hourly rate cannot be negative.")
        hourly_rate = float(input("Enter hourly rate: "))

    hours_worked = float(input("Enter total hours worked this month: "))
    while hours_worked < 0:
        print("❌ Hours worked cannot be negative.")
        hours_worked = float(input("Enter total hours worked this month: "))

    employee = Employee(name, hourly_rate, hours_worked)
    employees.append(employee)

print("\n==============================")
print("📄 PAYROLL SUMMARY")
print("==============================")

for emp in employees:
    print(emp.info())
    print("------------------------------")
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def reduce_stock(self, quantity):
        if quantity > self.stock:
            return False
        self.stock -= quantity
        return True

    def increase_stock(self, quantity):
        self.stock += quantity


class ShoppingCart:
    def __init__(self):
        self.items = {}  # {product_name: [product_object, quantity]}

    def add_item(self, product, quantity):
        if quantity <= 0:
            return "❌ Invalid quantity."

        if quantity > product.stock:
            return f"❌ Not enough stock for '{product.name}'. Available: {product.stock}"

        product.reduce_stock(quantity)

        if product.name in self.items:
            self.items[product.name][1] += quantity
        else:
            self.items[product.name] = [product, quantity]

        return f"✅ Added {quantity} x {product.name} to cart."

    def remove_item(self, product_name, quantity):
        if product_name not in self.items:
            return f"❌ '{product_name}' is not in the cart."

        product, current_qty = self.items[product_name]

        if quantity <= 0:
            return "❌ Invalid quantity."

        if quantity > current_qty:
            return f"❌ You only have {current_qty} of '{product_name}' in the cart."

        # Return stock
        product.increase_stock(quantity)

        # Update or remove item
        if quantity == current_qty:
            del self.items[product_name]
        else:
            self.items[product_name][1] -= quantity

        return f"🗑 Removed {quantity} x {product_name} from cart."

    def view_cart(self):
        if not self.items:
            return "🛒 Cart is empty."

        output = "\n🛒 Your Cart:\n"
        total = 0

        for product_name, (product, qty) in self.items.items():
            subtotal = product.price * qty
            total += subtotal
            output += f"- {product_name} (x{qty}) — ₱{subtotal:,.2f}\n"

        output += f"\n💰 Total Price: ₱{total:,.2f}"
        return output


# ---------------------------
# PROGRAM EXECUTION
# ---------------------------

# Predefined product list
products = {
    "Laptop": Product("Laptop", 45000, 5),
    "Mouse": Product("Mouse", 500, 20),
    "Keyboard": Product("Keyboard", 1500, 10),
    "Headset": Product("Headset", 2000, 8)
}

cart = ShoppingCart()

def show_products():
    print("\n📦 Available Products:")
    for name, product in products.items():
        print(f"- {name}: ₱{product.price} | Stock: {product.stock}")

while True:
    print("\n========== ONLINE SHOPPING CART ==========")
    print("1. View Products")
    print("2. Add to Cart")
    print("3. Remove from Cart")
    print("4. View Cart")
    print("5. Checkout & Exit")

    choice = input("Select an option: ")

    if choice == "1":
        show_products()

    elif choice == "2":
        show_products()
        product_name = input("Enter product name to add: ")

        if product_name not in products:
            print("❌ Invalid product selection.")
            continue

        quantity = int(input("Enter quantity: "))
        print(cart.add_item(products[product_name], quantity))

    elif choice == "3":
        product_name = input("Enter product name to remove: ")
        quantity = int(input("Enter quantity to remove: "))
        print(cart.remove_item(product_name, quantity))

    elif choice == "4":
        print(cart.view_cart())

    elif choice == "5":
        print("\n🛒 Final Cart Summary:")
        print(cart.view_cart())
        print("\nThank you for shopping!")
        break

    else:
        print("❌ Invalid option. Try again.")
```
